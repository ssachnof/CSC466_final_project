### get the total time of possession by the offense

```python
df = df.set_index(['game_id'])
df = df.query('play_type != "kickoff" and play_type != "no_play" and play_type != "extra_point"')
df = df.sort_values(by='game_seconds_remaining', ascending=False)
df
```

```python
df = df[['play_id', 'posteam', 'half_seconds_remaining', 'game_seconds_remaining', 'game_half']]
# df = df[df['game_half'] == 'Half1']
df = df.dropna()
df = df.reset_index()
df = df.sort_values(by=['game_id', 'game_seconds_remaining'], ascending=[True, False])
df = df.set_index('game_id')
df
```

```python
def get_half_top(half, teams, is_ot=False):
    pos_team = None
    tops = dict.fromkeys(teams, 0.0)
    pos_start = None
    half = half.reset_index(drop=True)
    
    
    for i, play in half.iterrows():
        if pos_team is None:
            pos_start = play.loc['half_seconds_remaining']
            pos_team = play.loc['posteam']
        elif play.loc['posteam'] != pos_team:#the ball possession somehow switched
            pos_end = play.loc['half_seconds_remaining']
            tops[pos_team] += pos_start - pos_end
            
            pos_start = pos_end
            pos_team = play.loc['posteam']
    # note that for ot, if the game ends in a tie, the last play in terms of top will be ignored
    # will come back to this if extra time at end of quarter
    
    if not is_ot:
        i = len(half) - 1
        tops[pos_team] += half['half_seconds_remaining'].loc[i]
    tops = pd.DataFrame(tops.items())
    tops.columns = ['team', 'time_of_possession']
    tops = tops.set_index('team')
    return tops



def get_game_top(game_id, df):
    game = df.loc[game_id]
    game = game.reset_index()
    teams = game['posteam'].unique()
    game = game.drop(columns=['game_id'])
    h1 = game[game['game_half'] == 'Half1']
    h1 = h1.sort_values(by= 'game_seconds_remaining', ascending=False)
    h2 = game[game['game_half'] == 'Half2']
    h2 = h2.sort_values(by= 'game_seconds_remaining', ascending=False)
    ot = game[game['game_half'] == "Overtime"]
    ot = ot.sort_values(by= 'game_seconds_remaining', ascending=False)
    h1_top = get_half_top(h1, teams)
    h2_top = get_half_top(h2, teams)
    ot_top = get_half_top(ot, teams, is_ot=True)
    game_top = h1_top + h2_top + ot_top#this is ok b/c pandas correctly lines up the indices
#     game_top = h1_top
    return game_top


    
    
```

```python
games = np.unique(df.index)
top_df = pd.DataFrame(columns=['game_id', 'team', 'time_of_possession'])
for game_id in games:
    game_df = get_game_top(game_id, df)
    game_df['game_id'] = game_id
    game_df = game_df.reset_index()
    game_df = game_df[['game_id', 'team', 'time_of_possession']]
    top_df = top_df.append(game_df, ignore_index=True)
top_df = top_df.set_index(['game_id', 'team'])
top_df = top_df.reset_index()
top_df
```

```python
#calculate completion percentage
qbr = df[df['play_type'] == 'pass']
qbr = qbr[['game_id', 'posteam','complete_pass']]
qbr = qbr.dropna()
qbr['num_attempts'] = qbr.groupby(['game_id', 'posteam'])['complete_pass']\
.transform(lambda x: x.count())
qbr['completions'] = qbr.groupby(['game_id', 'posteam'])['complete_pass'].transform(np.sum)
qbr = qbr.drop(columns=['complete_pass'])
qbr = qbr.drop_duplicates()
qbr['completion_ratio'] = qbr['completions'] / qbr['num_attempts']
qbr
```
